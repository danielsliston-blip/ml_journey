```python
# Day 01 â€” Setup sanity check

import sys, platform
print("Python:", sys.version)
print("Platform:", platform.platform())

# Core libraries you'll use constantly
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("NumPy:", np.__version__)
print("Pandas:", pd.__version__)

# Create a tiny dataset
df = pd.DataFrame({
    "day": [1, 2, 3, 4, 5],
    "minutes_studied": [90, 90, 90, 90, 90],
    "focus_score": [6, 7, 6, 8, 7]
})
df
```

```python
# Save + load (basic data pipeline skill)
csv_path = "day01_data.csv"
df.to_csv(csv_path, index=False)

df2 = pd.read_csv(csv_path)
df2
```

```python
# Simple plot (you'll do this constantly in ML work)
plt.figure()
plt.plot(df2["day"], df2["focus_score"], marker="o")
plt.title("Focus Score Over Days")
plt.xlabel("Day")
plt.ylabel("Focus Score")
plt.show()
```
